```python
pop_density = pd.read_excel("C:\\hongjaeyeon0\\html_css\\Team1TestProject\\hjy\\시군구_인구밀도.xlsx")
pop_density
```

```python
pop_density = pop_density.drop([0])
pop_density = pop_density.set_index("행정구역별")
```

```python
pop_density = pop_density.drop(["전국"])
pop_density
```

```python
pop_density = pop_density.loc[:, "2016.2":]
pop_density
```

```python
pop_density = pop_density.loc[:, ['2016.2', '2017.2', '2018.2', '2019.2', '2020.2', '2021.2']]
pop_density
```

```python
pop_density.rename(index = {'제주도' : '제주특별자치도'}, inplace=True)
pop_density
```

```python
pop_density = pop_density.stack()
pop_density
```

```python
merged_series = pd.concat([birthrate, pop_density], axis = 1)
merged_series
```

```python
merged_series.rename(columns={0:'합계출산율 증감율', 1:'인구밀도 증감율'},
                    inplace=True)
merged_series
```
